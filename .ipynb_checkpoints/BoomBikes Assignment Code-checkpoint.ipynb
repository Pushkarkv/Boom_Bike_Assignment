```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
df=pd.read_csv('day.csv')
```

```python
df.head()
```

```python
df.describe()
```

**Let's see if there are any null values.**

```python
df.info()
```

**Taget Variable is 'cnt'. But there are to other dependant variables that make it up.**

```python
x=(df['cnt']==df['casual']+df['registered'])
```

```python
x.value_counts() #We can see that all the values add up to make 'cnt' as expalined in the problem statement.
```

**Let us drop unwanted, 'casual' and 'registered' columns as they are not independant and we are supposed to predict 'cnt'**

```python
df.drop(columns=['instant', 'casual', 'registered'], inplace=True)
```

```python
df.head()
```

```python
df.dtypes
```

**Let us re-extract the year month dayofweek from the date and drop the redundant columns**

```python
df['date']=pd.to_datetime(df.dteday).dt.date
df['year']=pd.to_datetime(df.date).dt.year
df['month']=pd.to_datetime(df.date).dt.month_name()
df['dayofweek']=pd.to_datetime(df.date).dt.day_name()
```

```python
df.head()
```

```python
df.drop(columns=['dteday', 'yr', 'mnth', 'weekday', 'date'], inplace=True)
```

```python
df.head()
```

```python
df.describe()
```

**Seems like there are no outliers and all the quaters are resonably separated from each other**

```python
df['season'].value_counts()
```

```python
#Let us now try to map the numbers with then one-hot encode wherever there are more that two values and they are not ordinal.

df['season']=df['season'].map({1:'spring', 2:'summer', 3:'fall', 4:'winter'})
```

```python
df['weathersit'].value_counts() #Even tough we were told that there can be 4 values here we can see that only 3 siuations are available in data
```

```python
df['weathersit']=df['weathersit'].map({1:'Clear', 2:'Mist', 3:'Li_sn_rn', 4:'He_sn_rn'})
```

```python
#One hot encoding non ordinal categorical values

df=pd.get_dummies(df, columns=['season', 'weathersit', 'month', 'dayofweek'], drop_first=True)
```

```python
df.head()
```

```python
df.columns
```

```python
#Let us now see if there are any highly corelated fields.
sns.set(rc={'figure.figsize':(20,20)})
sns.heatmap(df.corr(), cmap='coolwarm', annot=True)
plt.show()
```

**'temp' and 'atemp' are highly correlated. No other fields have very high correlations.**

Let us drop temp as we know people won't normally look at the temparature to decide to ride, while how the temparature feels like might play major role if anyone is to go riding a bike.

```python
df.drop(columns=['temp'], inplace=True)
```

```python
sns.heatmap(df.corr(), cmap='coolwarm', annot=True)
plt.show()
```

```python
df.columns
```

**Now we are good to go, we can start to split data into train and test splits**

```python
df_train, df_test = train_test_split(df, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
df_train.shape
```

```python
df_test.shape
```

```python
#Let us scale independent numeric variables (except one hot encoded values) to increse the speed of training by making the gradient descent convergence faster.
num_features = ['atemp', 'hum','windspeed', 'year']
scaler = MinMaxScaler()
df_train[num_features] = scaler.fit_transform(df_train[num_features])
```

```python
df_train.head()
```

```python
y_train = df_train.pop('cnt')
X_train = df_train
```

**Let us Train our first model**

```python
X_train_wc = sm.add_constant(X_train)

lr = sm.OLS(y_train, X_train_wc).fit()
```

```python
lr.summary()
```

**Findings**

1. Clearly we can see that r_squared and adjusted are very comarable and high so the model is able to explain lot if dependant variable variance.
2. Probability of F statistic is very small meaning the overall model is also good fit.
3. We can see many of the variables like dayofweek, and month are not that significant as shown by thier p values.
4. The coefficients for the variables are also very large which could mean that the model might be unstable and if there are any collenierity the explainablity might get affected and caefficients can easily flip.

**Let us have a look at VIFs for the model**

```python
vif = pd.DataFrame()
vif['Features'] = X_train.columns
vif['VIF'] = [variance_inflation_factor(X_train.values, i) for i in range(X_train.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

**Findings**

1. Model is showing to have very high VIFs for hum and atem, other variables are having a lot better VIFs.

**Possible Actions**

1. I do believe that we can use the model as is if we just need to make predictions but as our use main goal is to get some clarity to the teams so that they can make better decisions to increase their profits, which can be done by increasing explainabilty. (by identifying important features that the team can improve to increase the profits)
2. I do believe that the we should not drop **hum** and **atemp** as they seem to be having very high significance as shown by their p values.
3. We should start by dropping **less significant features** that have **high VIF** as well, one at a time, monitoring the r_squared and updates to p values and VIFs.
4. But, clearly there are a lot of features (27 to be precise) and we could first try to use **Recursive Feature Elemination** to arrive at 12-15 features then use the manual techniques to drop further if needed.

**RFE**

```python
lm = LinearRegression()
lm.fit(X_train, y_train)

rfe = RFE(estimator=lm, n_features_to_select=12)
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
top_12_features = X_train.columns[rfe.support_]
top_12_features
```

```python
X_train.columns[~rfe.support_]
```

**Findings**

1. As we were expecting majority of the dayofweeks and months do not play a significant role to predict the response variable and hence have been removed.
2. Apart from the above features workingday also does not play major role in predicting the reponse variable.
3. All the seasons are kept meaning they are relatively significant.
4. Even tough we had seen very high VIFs for atemp and hum they play a significant role in prediction accoring to RFE.

**Let us now see how the updated model performs using OLS**

```python
X_train_rfe = X_train[top_12_features] #Keeping top 12 features as found in RFE.
X_train_wc = sm.add_constant(X_train_rfe)

lr = sm.OLS(y_train, X_train_wc).fit()
```

```python
lr.summary()
```

**Findings**

1. Even after dropping 15 features we can see that R_squared has remained 0.831 and we are able to explain a lot varience in response variable. Infact Ajusted_R_squared has remained same and clearly we should prefter this model when compared to the one built before.
2. The probability of F-statistic has also decreased meaning the new model is overall better fit than than before.
3. All the p values associacted with each indpendant variables now have very small values which is an improvement from previous model.
4. The coefficients are still large meaning the model might be ovefitting, we can try to see if we can drop any more features and still get similar R_squared score.
5. We can start by droping season_summer but before doing that let us try check the updated VIFs.

```python
vif = pd.DataFrame()
vif['Features'] = X_train_rfe.columns
vif['VIF'] = [variance_inflation_factor(X_train_rfe.values, i) for i in range(X_train_rfe.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

**Findings**

1. Model is still showing very high VIFs for hum and atem, other variables are having a lot better VIFs.

**Possible Actions**

1. Maybe the two variables hum and atemp are collinear, let us try to drop hum and see if it improves, and keeps similar R_squred.

```python
X_train_rfe.pop('hum')
```

```python
X_train_wc = sm.add_constant(X_train_rfe)

lr = sm.OLS(y_train, X_train_wc).fit()
```

```python
lr.summary()
```

```python
vif = pd.DataFrame()
vif['Features'] = X_train_rfe.columns
vif['VIF'] = [variance_inflation_factor(X_train_rfe.values, i) for i in range(X_train_rfe.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

**Findings**

1. After dropping hum also the R_squred and Ajusted_R_suared did not see any significat dip whiich means that the new model is definitly better fit.
2. VIF for atemp has also reduced below 5 which is great.
3. Slight increase in p value us season_summer, we can drop it and see if there is and significant change in overall R_squared.

```python
X_train_rfe.pop('season_summer')

X_train_wc = sm.add_constant(X_train_rfe)

lr = sm.OLS(y_train, X_train_wc).fit()
```

```python
lr.summary()
```

```python
vif = pd.DataFrame()
vif['Features'] = X_train_rfe.columns
vif['VIF'] = [variance_inflation_factor(X_train_rfe.values, i) for i in range(X_train_rfe.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

**Findings**

1. After dropping season_summer also the R_squred and Ajusted_R_suared did not see any significat dip which means that the new model is definitly good fit.
2. VIFs have also improved a lot.
3. All the p values also have become low meaning the new model has the variables that are significant witch is great.

**I think the new model is great fit and let us do Residual Analysis now**

```python
y_train_pred = lr.predict(X_train_wc)
```

```python
sns.set(rc={'figure.figsize':(8,8)})
sns.distplot((y_train - y_train_pred))
```

```python
sum(y_train - y_train_pred) #Sum of residuals
```

```python
sum(y_train - y_train_pred)/len(y_train_pred) #Mean of residuals
```

```python
fig = plt.figure()
plt.scatter(y_train - y_train_pred, y_train_pred) #To test for Homoscedasticity
fig.suptitle('residual vs y_pred', fontsize = 20)              # Plot heading 
plt.xlabel('residual', fontsize = 18)                          # X-label
plt.ylabel('y_pred', fontsize = 16)
```

**Findings**

1. Clearly the errors terms are normally distributed with zero at center.
2. The sum and mean of residual are also very close to zero.
3. The residuals don't seem to follow any patterns as shown by the scatter plot residuals and predicted values. Meaning the data is **heteroscedastic**.

**Let us now try to make predictions using the final model.**

```python
#Making test data ready for making prediction.

num_features = ['atemp', 'hum','windspeed', 'year']
df_test[num_features] = scaler.transform(df_test[num_features])
```

```python
y_test=df_test.pop('cnt') #popping the dependent variable and creating y
```

```python
train_list=X_train_rfe.columns.tolist() #assigning training variables to a list
```

```python
X_test=df_test[train_list]  # Using the above list to filter columns used for training the model

X_test_wc = sm.add_constant(X_test) #Adding constant value for using OLS
```

```python
y_pred = lr.predict(X_test_wc) #Predicting values using test data
```

**Model Evaluation**

```python
# Plotting y_test and y_pred to understand the spread

fig = plt.figure()
plt.scatter(y_test, y_pred)
fig.suptitle('y_test vs y_pred', fontsize = 20)              # Plot heading 
plt.xlabel('y_test', fontsize = 18)                          # X-label
plt.ylabel('y_pred', fontsize = 16) 
```

```python
r2_score(y_test, y_pred) #Testing R_squred
```

```python
r2_score(y_train,y_train_pred) #Training R_squred
```
